# Barplot

```python
# Bar plot of Match type
a=ds.matchType.value_counts().plot.bar()
a.set_xlabel('Match Type')
a.set_ylabel('Frequency')
a.set_title('Barplot of Match type')
```

```python
a=ds.headshotKills.value_counts().plot.bar()
a.set_xlabel('Headshot Kills')
a.set_ylabel('Frequency')
a.set_title('Barplot of Headshot Kills')
```
